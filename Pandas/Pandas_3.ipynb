## Stacking and Unstacking 
Multi Indexing in Rows <br>
Closely related to the pivot() method are the related stack() and unstack() methods available on Series and DataFrame. These methods are designed to work together with MultiIndex objects (see the section on hierarchical indexing). Here are essentially what these methods do:<br>

stack: pivot a level of the (possibly hierarchical) column labels, returning a DataFrame with an index with a new inner-most level of row labels.<br>

unstack: (inverse operation of stack) pivot a level of the (possibly hierarchical) row index to the column axis, producing a reshaped DataFrame with a new inner-most level of column labels.<br>

## Apply and Transform *** (Very Important Interview point of View)

### Apply 

One alternative to using a loop to iterate over a DataFrame is to use the pandas apply() method.<br> 
This function acts as a map() function in Python. It takes a function as an input and applies this function to an entire DataFrame.<br>
If you are working with tabular data, you must specify an axis you want your function to act on (0 for columns; and 1 for rows).

```python
df = pd.DataFrame({'A': [1,2,3], 'B': [10,20,30] })
df
```

```python
def plus_10(x):
    return x+10
```

```python
df = df.apply(plus_10) # df.apply(lambda x: x+10)
df
```

```python
df.apply(lambda x:x.sum())
```

### Transform

It is similar to Apply

One of the most compelling usages of Pandas transform() is combining grouby() results.

```python
df = pd.DataFrame({
  'restaurant_id': [101,102,103,104,105,106,107],
  'address': ['A','B','C','D', 'E', 'F', 'G'],
  'city': ['London','London','London','Oxford','Oxford', 'Durham', 'Durham'],
  'sales': [10,500,48,12,21,22,14]
})
```

```python
new_trans = df
new_trans['city_total_sales'] = df.groupby('city')['sales'].transform(sum)
new_trans    
```

```python
# to do the same operation in Apply
city_sales = df.groupby('city')['sales'].apply(sum).rename('city_total_sales').reset_index()
print(city_sales)   
df_new = pd.merge(df, city_sales, how='left')
df_new
```

<b> What are the differences Apply and  Transform? </b><br>
Here are the three main differences<br>
<ul>
<li>(1) transform() works with function, a string function, a list of functions, and a dict. However, apply() is only allowed with function.</li>
<li>(2) transform() cannot produce aggregated results.</li>
<li>(3) apply() works with multiple Series at a time. But, transform() is only allowed to work with a single Series at a time.</li>
</ul>
