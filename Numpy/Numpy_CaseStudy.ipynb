```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
astronaut = data.astronaut()
```

```python
type(astronaut)
```

```python
astronaut.size
```

```python
plt.imshow(astronaut)
```

```python
a = astronaut[100:, 50:, :]
plt.imshow(a)
plt.show()
```

```python
a1,a2 =np.split(astronaut, 2)
plt.imshow(a1)
plt.show()
plt.imshow(a2)
plt.show()
```

```python
cropped_astronaut = astronaut[:,:,:]
b1,b2 = np.split(cropped_astronaut, 2, axis = 1)
plt.imshow(b1)
plt.show()
plt.imshow(b2)
plt.show()
```

```python
plt.imshow(np.concatenate((a1,a2)))
plt.show()
plt.imshow(np.concatenate((b1,b2)))
plt.show()
plt.imshow(np.concatenate((b1,b2), axis = 1))
plt.show()
```
